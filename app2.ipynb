```python

#this is the best code for image recognization

import cv2
import numpy as np
from tensorflow.keras.models import load_model


categories = ['Comedor', 'Gamer', 'Mesedoras', 'Taburete', 'Terraza']

model_path = 'D:\\chair\\chair detection model cv\\chairvgg16.h5'  # Replace with the path to your model
model = load_model(model_path)

def preprocess_image(image, imgsize=224):
    """Preprocess the image to the required input shape for the model."""
    img = cv2.resize(image, (imgsize, imgsize))
    img = img / 255.0  # Normalize the image
    img = np.expand_dims(img, axis=0)  # Expand dimensions to match the model input
    return img

def predict_category(image, model, categories):
    """Predict the category of the image."""
    img = preprocess_image(image)
    prediction = model.predict(img)
    predicted_index = np.argmax(prediction, axis=1)[0]
    confidence = prediction[0][predicted_index]
    
    if confidence < 0.8:
        return "No Chair", confidence
    else:
        return categories[predicted_index], confidence

def draw_rectangle(frame, predicted_category):
    """Draw a rectangle around the frame and display the predicted category."""
    height, width, _ = frame.shape

    cv2.rectangle(frame, (0, 0), (width, height), (0, 255, 0), 2)
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame, predicted_category, (10, height - 10), font, 1, (0, 255, 0), 2, cv2.LINE_AA)
    cv2.imshow('Chair Detection', frame)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    predicted_category, confidence = predict_category(frame, model, categories)


    draw_rectangle(frame, predicted_category)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
